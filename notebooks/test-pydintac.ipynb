```python
from pydantic import BaseModel
from dataclasses import dataclass
```

```python
class Person(BaseModel):
  name:str
  age:int
  city: str 
person  = Person(name="muhammed", age=23, city='benha')
print(person)
print(type(person))
```

```python
@dataclass
class Person():
  name:str
  age:int
  city:str

  
person  = Person(name="muhammed", age=23, city='benha')
print(person)
print(type(person))
```

```python
person1= Person(name="muhammed", age=23, city=1)
print(person1)
```

```python
from typing import Optional

class Employee(BaseModel):
  id:int 
  name: str
  department:str
  salary: Optional[float] = None
  is_active: Optional[bool] = True
```

```python
emp = Employee(id=1, name='mohammed',department='SD')
print(emp)
```

```python
emp1 = Employee(id=2, name='ahemd',department='SD',salary=20000)
print(emp1)
```

```python
from typing import List
class Classroom(BaseModel):
  room_number:str
  students:List[str]
  capacity:int
```

```python
classroom = Classroom(room_number='1',students=['ahmed','mohammed'],capacity=2)
print(classroom)
```

```python
class Address(BaseModel):
  street:str
  city:str
  zip_code:str
  
class Customer(BaseModel):
  customer_id:int
  name: str
  address:Address
```

```python
customer = Customer(customer_id=1, name='john',
                    address=Address(
                      street='tes',
                      city='benha',
                      zip_code='1233'
                    ))
print(customer)
```

```python
customer = Customer(customer_id=1, name='john',
                    address={
                      'street':'tes',
                      'city':'benha',
                      'zip_code':'1233'
                    })
print(customer)
```

```python
from pydantic import Field
class Item(BaseModel):
  name:str=Field(min_length=2, max_length=50)
  price:float = Field(ge=0, le=1000, description='item price')
  quantity: int = Field(ge=0)
  
```

```python
class User(BaseModel):
  username:str = Field(..., description='Unique username for user')
  age:int = Field(default=18, description='user age, defaults is 18')
  email:str = Field(default_factory=lambda:"user@example.com", description="default email address")
```

```python
print(User.model_json_schema())
```
